```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import seaborn as sb

style.use('seaborn-talk')

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
dataset = pd.read_csv("C:/Users/auteu/Desktop/Projects/UOS/WaRaConv/ProcessedDataSet/서울시 생활인구_주민등록인구.csv", encoding='cp949',engine='python')
dataset
X = dataset[["2020년08월_총인구수","8월평균총생활인구수"]]
X
```

```python
from sklearn.cluster import KMeans

init_centroid = 'k-means++'

km = KMeans(n_clusters=3,init=init_centroid, random_state=0)
y_km=km.fit_predict(X)

dataset['클러스터']=km.labels_
dataset
```

```python
sb.lmplot(x='2020년08월_총인구수',y='8월평균총생활인구수',fit_reg=False,data=dataset,hue="클러스터")
plt.title('생활인구-거주인구')
plt.savefig('clustered1.png')
```
